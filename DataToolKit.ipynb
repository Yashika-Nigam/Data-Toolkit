1. What is NumPy, and why is it widely used in Python?

NumPy is a powerful Python library used for numerical and scientific computing. It provides multi-dimensional arrays, mathematical functions, vectorized operations, and tools for linear algebra and statistics. NumPy is widely used because it is extremely fast, memory-efficient, and forms the foundation of many libraries like Pandas, Scikit-learn, and TensorFlow.

2. How does broadcasting work in NumPy?

Broadcasting allows NumPy to perform arithmetic operations on arrays of different shapes. Instead of resizing smaller arrays, NumPy “stretches” them virtually so operations apply element-wise. It simplifies mathematical operations and improves performance without needing explicit loops.

3. What is a Pandas DataFrame?

A Pandas DataFrame is a two-dimensional, tabular data structure with rows and columns. It is similar to an Excel sheet and allows easy data cleaning, manipulation, analysis, and transformation. It supports heterogeneous data and is one of the most widely used structures in data analysis.

4. Explain the use of the groupby() method in Pandas.

The groupby() function groups data based on one or more columns and allows aggregate operations like sum, mean, count, or custom aggregations. It is used to analyze grouped patterns, summarize data, and extract meaningful insights from datasets.

5. Why is Seaborn preferred for statistical visualizations?

Seaborn is preferred because it produces attractive, informative, and statistically meaningful plots with minimal code. It supports built-in themes, handles complex datasets easily, and integrates smoothly with Pandas DataFrames.

6. What are the differences between NumPy arrays and Python lists?

NumPy arrays are homogeneous, faster, and support vectorized operations, while Python lists are heterogeneous and slower. Arrays store data in contiguous memory blocks, enabling efficient mathematical computation. Lists don’t allow element-wise operations without loops.

7. What is a heatmap, and when should it be used?

A heatmap is a graphical representation of data values depicted using color intensity. It is used to visualize correlations, patterns, and high–low concentrations in data, especially in correlation matrices or frequency tables.

8. What does the term “vectorized operation” mean in NumPy?

Vectorized operations are computations applied directly to entire arrays without explicit loops. They make code faster, cleaner, and more efficient by leveraging optimized C-level implementations.

9. How does Matplotlib differ from Plotly?

Matplotlib is primarily used for static, publication-quality plots, while Plotly is used for interactive visualizations. Plotly supports zooming, tooltips, and web-based dashboards, whereas Matplotlib is more traditional and highly customizable.

10. What is the significance of hierarchical indexing in Pandas?

Hierarchical or MultiIndexing allows multiple levels of indexing in rows or columns. It helps represent higher-dimensional data, simplifies complex datasets, and enables more flexible data selection and aggregation.

11. What is the role of Seaborn’s pairplot() function?

pairplot() visualizes pairwise relationships between multiple numerical variables. It creates scatterplots for combinations of features and histograms on diagonals, helping analyze patterns, correlations, and distributions quickly.

12. What is the purpose of the describe() function in Pandas?

describe() generates summary statistics like mean, median, count, min, max, and quartiles for numerical columns. It provides a quick overall understanding of data distribution and central tendencies.

13. Why is handling missing data important in Pandas?

Missing data can distort analysis, affect model accuracy, and lead to incorrect conclusions. Pandas provides functions like isna(), fillna(), and dropna() to clean datasets and maintain data integrity.

14. What are the benefits of using Plotly for data visualization?

Plotly creates interactive, dynamic, and web-friendly charts. It supports 2D/3D plots, dashboards, and animations, making it ideal for storytelling, presentations, and exploratory analysis.

15. How does NumPy handle multidimensional arrays?

NumPy uses the ndarray object to store multi-dimensional data. It supports indexing, slicing, reshaping, broadcasting, and mathematical operations across any number of dimensions efficiently.

16. What is the role of Bokeh in data visualization?

Bokeh is used to create interactive and browser-based visualizations. It supports real-time dashboards, streaming data, and high-performance rendering, making it suitable for modern web applications.

17. Explain the difference between apply() and map() in Pandas.

map() works on Series and applies a function element-wise.
apply() works on both Series and DataFrames and can apply row-wise or column-wise operations. apply() is more flexible, while map() is simpler.

18. What are some advanced features of NumPy?

Advanced features include broadcasting, masked arrays, linear algebra (linalg), random sampling, Fourier transforms, memory mapping, and structured arrays for complex data manipulation

19. How does Pandas simplify time series analysis?

Pandas provides date-range generation, resampling, shifting, frequency conversion, time-based indexing, and built-in date parsing. These features make time-series handling extremely efficient.

20. What is the role of a pivot table in Pandas?

Pivot tables summarize data by reorganizing columns and applying aggregate functions. They help analyze trends, perform multi-level grouping, and structure data in meaningful ways.

21. Why is NumPy’s array slicing faster than Python’s list slicing?

NumPy arrays use contiguous memory and optimized C-based operations. Slicing creates views instead of copies, making it extremely fast. Python lists store references, making slicing slower.

22. What are some common use cases for Seaborn?

Common use cases include statistical plots like histograms, bar plots, violin plots, box plots, correlation heatmaps, jointplots, and regression plots.

```python
#ques-4
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Creating simple sample data
data = {
    'Height': [150, 160, 170, 165, 155],
    'Weight': [50, 60, 70, 65, 55],
    'Age':    [20, 25, 30, 22, 27]
}

df = pd.DataFrame(data)

# Correlation matrix
corr = df.corr()
print(corr)

# Heatmap
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.show()

```
